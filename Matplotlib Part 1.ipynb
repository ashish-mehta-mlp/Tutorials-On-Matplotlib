### Video 1: Introduction to Matplotlib

## Matplotlib

- Matplotlib is a graph plotting library in Python. 
- Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

**1. We are creating a plot for median salaries of developers by age**

```python
# creating a sample data
dev_x = range(25,36,1) # dev_x is a subset of age between 25 to 35
```

```python
dev_x1 = []
for x in dev_x:
    dev_x1.append(x)
```

```python
dev_x1 #dev_x1 is the subset of age between 25-35 and these are the values on our x axis
```

```python
# creating a list of median salaries between age 25 and 35. These are the values on the y-axis 
dev_y = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
```

- We can make the plot appear by using plt.plot()
- First we input the values for x-axis then we give the values for y-axis
- Values on x-axis are the age and values on y-axis are the median salaries

```python
plt.plot(dev_x1, dev_y)
```

- How to add some more information to the plot

```python
# adding title, xlabel and ylabel to the graph to make it more understandable

plt.plot(dev_x1, dev_y)
plt.title("Median Salary (USD) by Age")
plt.xlabel("Age")
plt.ylabel("Median Salary (USD)")
```

## Plotting two values on a single graph

**2. Comparing the median salaries of developers with median salaries of "Python Developers" based on age**

- The values on the x-axis i.e the subset of age will remain the same.
- Hence we will rename dev_x1 as age_x and use it for both the graphs as x-axis values

```python
# Common value of x-axis
age_x = []
age_x = dev_x1
```

```python
# Values of y-axis for normal developers
dev_y = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
```

```python
# Values of y-axis for python developers
pydev_y = [45372, 48476, 53850, 57287, 63016, 65998, 70003, 70000, 71496, 75370, 83640]
```

```python
plt.plot(age_x, dev_y)
plt.plot(age_x, pydev_y)
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
```

 - Now we are getting 2 lines. But we dont know which line is which. So we have to add LEGENDS
 - There are two ways to add legend

```python
# Technique 1: We can add a legend by passing a list in the order that they were added in the plot
plt.plot(age_x, dev_y)
plt.plot(age_x, pydev_y)
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend(["All_Developers", "Python_Developers"])
```

```python
# Technique 2: [Better technique] - Adding label parameter in plt.plot()
plt.plot(age_x, dev_y, label = "All_Developers")
plt.plot(age_x, pydev_y, label = "Python_Developers")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend() #plt.legend() takes label parameters in plt.plot as its arguments
```

- Formatting the Graph
  1. Changing Colour of the plot

- We can pass all the formatting commands as a single line of code but it becomes hard to read. 
- Passing all formatting code at once using fmt string

A format string consists of a part of color, marker and line

fmt = "[marker][Line][Color]"

```python
plt.plot(age_x, dev_y, "k--", label = "All_Developers") # k stands for black and -- is the line type
plt.plot(age_x, pydev_y, "b-",label = "Python_Developers") # b stands for blue and - is the line type
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
```

```python
# Making the code more readable
plt.plot(age_x, dev_y, color = "k", linestyle = "--", label = "All_Developers")
plt.plot(age_x, pydev_y, color = "b", linestyle = '-', label = "Python_Developers")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
```

```python
# Adding markers to your lines
plt.plot(age_x, dev_y, color = "k", linestyle = "--", marker = ".", label = "All_Developers")
plt.plot(age_x, pydev_y, color = "b", linestyle = '-', marker = "o",label = "Python_Developers")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
```

- Format string has very limited number of colours
- We can find large variation of colours online which are called hex-values
- The first 2 values after # is a shade of red, next 2 values is a shade of green, last 2 values are a shade of blue

```python
plt.plot(age_x, dev_y, color = "#444444", linestyle = "--", label = "All_Developers")
plt.plot(age_x, pydev_y, color = "#5a7d9a", linestyle = '-',label = "Python_Developers")
jsdev_y = [37810, 43515, 46823, 49293, 53437, 56373, 62375, 66674, 68745, 68746, 74583]
plt.plot(age_x, jsdev_y, color = "#adad3b", linestyle = '-',label = "Java_script_Developers")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
```

- Formatting the Graph
  2. Changing thickness of the line

- In the above graph we can see that the green line is overlapping the dotted line. This is because the dotted line was passed first in the code and then the other two lines were passed. Hence to make the dotted lines appear above green line, we need to pass the dotted line after green line

```python
# changing the width of python line and java_script line
plt.plot(age_x, pydev_y, color = "#5a7d9a", linestyle = '-',linewidth = 4, label = "Python_Developers")
jsdev_y = [37810, 43515, 46823, 49293, 53437, 56373, 62375, 66674, 68745, 68746, 74583]
plt.plot(age_x, jsdev_y, color = "#adad3b", linestyle = '-',linewidth = 4, label = "Java_script_Developers")
plt.plot(age_x, dev_y, color = "#444444", linestyle = "--", label = "All_Developers")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
```

```python
#add padding to graph'
plt.plot(age_x, pydev_y, color = "#5a7d9a", linestyle = '-',linewidth = 4, label = "Python_Developers")
jsdev_y = [37810, 43515, 46823, 49293, 53437, 56373, 62375, 66674, 68745, 68746, 74583]
plt.plot(age_x, jsdev_y, color = "#adad3b", linestyle = '-',linewidth = 4, label = "Java_script_Developers")
plt.plot(age_x, dev_y, color = "#444444", linestyle = "--", label = "All_Developers")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
plt.tight_layout() # To add padding to the graph

#Padding is the space between the elements and the 4 borders
```

```python
# adding grid to the plots
plt.plot(age_x, pydev_y, color = "#5a7d9a", linestyle = '-',linewidth = 4, label = "Python_Developers")
jsdev_y = [37810, 43515, 46823, 49293, 53437, 56373, 62375, 66674, 68745, 68746, 74583]
plt.plot(age_x, jsdev_y, color = "#adad3b", linestyle = '-',linewidth = 4, label = "Java_script_Developers")
plt.plot(age_x, dev_y, color = "#444444", linestyle = "--", label = "All_Developers")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
plt.tight_layout()
plt.grid(True)
```

### How to change the styles of a plot and make them look better

- Built-in styles of matplotlib

```python
# look at the styles that are available

plt.style.available
```

```python
# Using the 'fivethirtyeight' style
plt.plot(age_x, pydev_y, color = "#5a7d9a", linestyle = '-',linewidth = 4, label = "Python_Developers")
plt.style.use("fivethirtyeight")
jsdev_y = [37810, 43515, 46823, 49293, 53437, 56373, 62375, 66674, 68745, 68746, 74583]
plt.plot(age_x, jsdev_y, color = "#adad3b", linestyle = '-',linewidth = 4, label = "Java_script_Developers")
plt.plot(age_x, dev_y, color = "#444444", linestyle = "--", label = "All_Developers")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
plt.tight_layout()
plt.grid(True)
```

```python
# Removing the custom colors, widths and grid of the plot and using 'fivethirtyeight' style colors
plt.plot(age_x, pydev_y, linestyle = '-', label = "Python_Developers")
plt.style.use('fivethirtyeight')
jsdev_y = [37810, 43515, 46823, 49293, 53437, 56373, 62375, 66674, 68745, 68746, 74583]
plt.plot(age_x, jsdev_y, linestyle = '-', label = "Java_script_Developers")
plt.plot(age_x, dev_y, color = "#444444", linestyle = "--", label = "All_Developers")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
plt.tight_layout()
```

```python
# Using the 'ggplot' style
plt.plot(age_x, pydev_y, linestyle = '-', label = "Python_Developers")
plt.style.use('ggplot')
jsdev_y = [37810, 43515, 46823, 49293, 53437, 56373, 62375, 66674, 68745, 68746, 74583]
plt.plot(age_x, jsdev_y, linestyle = '-', label = "Java_script_Developers")
plt.plot(age_x, dev_y, color = "#444444", linestyle = "--", label = "All_Developers")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
plt.tight_layout()
```

```python
# Comic style graph: Remove the style method and add xkcd method 
plt.plot(age_x, pydev_y, linestyle = '-', label = "Python_Developers")
plt.xkcd()
jsdev_y = [37810, 43515, 46823, 49293, 53437, 56373, 62375, 66674, 68745, 68746, 74583]
plt.plot(age_x, jsdev_y, linestyle = '-', label = "Java_script_Developers")
plt.plot(age_x, dev_y, color = "#444444", linestyle = "--", label = "All_Developers")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
plt.tight_layout()

# looks like a handmade graph
```
