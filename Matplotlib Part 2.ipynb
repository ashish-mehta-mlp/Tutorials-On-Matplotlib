### Creating horizontal barcharts

- Horizontal barcharts are used when we have a large amount of data and it looks to crowded in a vertical plot
- We will be loading the data from a csv file

#### We have to create a bar-plot showcasing which is the most popular computing language the users have worked on

```python
data = pd.read_csv("C:/Users/Ashish/Desktop/Python Tutorials/CSV files/data.csv")
```

```python
data.head()
```

```python
ids = data["Responder_id"]
lang_responses = data["LanguagesWorkedWith"]
```

- We are making use of COUNTERS to keep track of the value_counts of the languages  

```python
from collections import Counter
```

```python
# instantiating the counter
lang_counter = Counter()
```

```python
# updating the counter
for response in lang_responses:
    lang_counter.update(response.split(";"))
```

```python
lang_counter # We can see that lang_counter is a dictionary with "name of language" as keys and "number of users" as values
```

```python
# seperating the keys and values as seperate lists
language = []
popularity = []

for item in lang_counter.most_common(15):
    language.append(item[0])
    popularity.append(item[1])
```

```python
print(language)
print(popularity)
```

```python
# Creating simple barplot first
plt.bar(language, popularity)
plt.title("Popularity of Programming languages")
plt.ylabel("Popularity")
plt.xlabel("Programming language")
```

```python
# In the above graph, we can see that the xlabels are not clear at all. Hence we make use of horizontal bar graph
```

```python
`#Creating simple bar plot (barh)
plt.barh(language, popularity)
plt.title("Popularity of Programming languages")
plt.xlabel("Popularity")
plt.ylabel("Programming language")
```

```python
# We can reverse the bar graph by keeping the most popular language on the top
language.reverse()
popularity.reverse()
plt.barh(language, popularity)
plt.title("Popularity of Programming languages")
plt.xlabel("Popularity")
plt.ylabel("Programming language")
```

### Video 3: Pie Charts

- In pie chart, instead of plotting certain values for each category, it shows us the proportion that each category contributes to a whole.
- As the name implies, we can think of the chart as a pie and each category will be a slice of that pie.
- Larger the size of the category, larger the share it has in the pie

```python
# Creating a simple Pie chart

import matplotlib.pyplot as plt
import pandas as pd
plt.style.use("fivethirtyeight")
```

```python
slices = [60,20,20]
plt.pie(slices)
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

## Adding labels to pie chart

- labels are used to indicate what does each slice represent

```python
slices = [60,20,20]
labels = ["Sixty", "Twenty", "Twenty"] #The labels have to be in the same order as the slices.  
plt.pie(slices, labels = labels)
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

- It is not necessary that the values in the slices should add up to 100. The code automatically adjusts the values to fit in the pie

```python
slices = [120,60,40]
labels = ["One hundred and twenty", "Sixty", "Forty"] #The labels have to be in the same order as the slices.  
plt.pie(slices, labels = labels)
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
# We can create seperators on the edges of each slice using [wedgeprops]
slices = [60,40,20]
labels = ["Sixty", "Forty", "Twenty"]
plt.pie(slices, labels = labels, wedgeprops = {'edgecolor': 'black'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

- Specifying colours for each slices

To change the colour of the slices, all we need to do is create list of colours and pass it to plt.pie()

```python
slices = [60,40,20,15]
labels = ["Sixty", "Forty", "Twenty", "Fifteen"]
colors = ["blue", "red", "yellow", "green"]
plt.pie(slices, labels = labels, colors = colors, wedgeprops = {'edgecolor': 'black'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

- The above colours look like elementary school colours
- Hence we make use of hex colours

```python
slices = [60,40,20,15]
labels = ["Sixty", "Forty", "Twenty", "Fifteen"]
colors = ["#008fd5", "#fc4f30", "#e5ae37", "#6d904f"]
plt.pie(slices, labels = labels, colors = colors, wedgeprops = {'edgecolor': 'black'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

- This looks like a more professional chart now

```python
## Now we will use the top 5 languages to draw the piechart

import matplotlib.pyplot as plt
plt.style.use("ggplot")
slices = [59219, 55466, 47544, 36443, 35917]
labels = ["Javascript", "HTML/CSS", "SQL", "Python", "Java"]
plt.pie(slices, labels = labels, wedgeprops = {"edgecolor":"black"})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
## Putting emphasis on one single slice. (eg: python slice)
## To do this we will pass an explode argument which is a list of values that will offset the slice

import matplotlib.pyplot as plt
plt.style.use("ggplot")
slices = [59219, 55466, 47544, 36443, 35917]
labels = ["Javascript", "HTML/CSS", "SQL", "Python", "Java"]
explode = [0,0,0,0.2,0] # The zeros are for slices that we dont want to explode. and 0.2 is for python slice in the order of labels
plt.pie(slices, labels = labels, explode = explode, wedgeprops = {"edgecolor":"black"})
plt.title("My Awesome Pie Chart")
plt.tight_layout()

# The 0.2 in the explode means seperate the given slice by 20% of the radius 
```

```python
# Improving the asthetics of the chart: shadow = True

import matplotlib.pyplot as plt
plt.style.use("ggplot")
slices = [59219, 55466, 47544, 36443, 35917]
labels = ["Javascript", "HTML/CSS", "SQL", "Python", "Java"]
explode = [0,0,0,0.2,0] 
plt.pie(slices, labels = labels, explode = explode, shadow = True, wedgeprops = {"edgecolor":"black"})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
# We can rotate the chart as per our wish. (startangle)
import matplotlib.pyplot as plt
plt.style.use("ggplot")
slices = [59219, 55466, 47544, 36443, 35917]
labels = ["Javascript", "HTML/CSS", "SQL", "Python", "Java"]
explode = [0,0,0,0.2,0] 
plt.pie(slices, labels = labels, explode = explode, shadow = True, startangle = 90, wedgeprops = {"edgecolor":"black"})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
# Display the percentage that the each slice is occupying in the pie: (autopct = "%1.1f%%") autopct = auto percent

import matplotlib.pyplot as plt
plt.style.use("ggplot")
slices = [59219, 55466, 47544, 36443, 35917]
labels = ["Javascript", "HTML/CSS", "SQL", "Python", "Java"]
explode = [0,0,0,0.2,0] 
plt.pie(slices, labels = labels, explode = explode, shadow = True, startangle = 90, autopct = "%1.1f%%", wedgeprops = {"edgecolor":"black"})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```
