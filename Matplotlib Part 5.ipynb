## Video 9: Plotting Live data in Real Time

- We will be plotting data that is being generated real time
- Real time data is data that is changing frequently and we want to keep a track of it

```python
# Plotting a simple plot

import random
import pandas as pd
import matplotlib.pyplot as plt
from itertools import count
```

```python
plt.style.use("ggplot")
x = [0,1,2,3,4,5]
y = [0,3,4,5,1,2]
plt.plot(x,y, marker = "o")
plt.tight_layout()
```

- count() function from itertools
- By default, its use is to count one number at a time each time we get a new value.

```python
# Ploting live incoming data 
from matplotlib.animation import FuncAnimation
plt.style.use("ggplot")
x = []
y = []

index = count()

def animate(i):
    x.append(next(index))
    y.append(random.randint(0,5))
    plt.cla() # cla means clear axis
    plt.plot(x,y)

ani = FuncAnimation(plt.gcf(), animate, interval = 1000)
plt.tight_layout()

# In order to plot a live data, we have to create a function and store the values of x and y in it. Then we have to run the function 
# Lets say we want to run this animate function every second and plot these values that are getting appended to our list x & y
# We will plot this real time data using funcanimate class from matplotlib animate module.
# We will ask funcanimation to run on a specific interval, so that we can plot that incoming data


    
```

- In funcAnimation we pass the following arguments:
1. First we pass the figure we want to animate.
2. Next we pass the function we want to run for our animation
3. We give the interval of plotting. The interval is in miliseconds. 1000 miliseconds = 1 second


- The colour of the plot keeps on changing in the above plot.
- This is because, after every second new plot is drawn and it is overlapping on the previous plot.
- One way to solve this problem is to clear the axis.
- When we clear the axis, it deletes the previous line-plots

```python
import random
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []

index = count()


def animate(i):
    data = pd.read_csv('C:/Users/Ashish/Desktop/Python Tutorials/data.csv')
    x = data['x_value']
    y1 = data['total_1']
    y2 = data['total_2']

    plt.cla()

    plt.plot(x, y1, label='Channel 1')
    plt.plot(x, y2, label='Channel 2')

    plt.legend(loc='upper left')
    plt.tight_layout()


ani = FuncAnimation(plt.gcf(), animate, interval=1000)

plt.tight_layout()
```
