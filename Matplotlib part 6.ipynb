Q. What are figure and axis?

Ans: - Figure means the **container (box)** in which the plot is drawn 
     - **Axis means the plot**
        
In the above graph we can see that there is 1 figure and 1 axis. We can change the number of figures or axis to bifurcate the plot as per our choice.

For this, we make use of subplot() method

```python
# Creating axis and figure object
ages = data["Age"]
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]

fig, ax = plt.subplots()

plt.plot(ages, py_salaries, label = "Python")
plt.plot(ages, js_salaries, label = "JavaScript")
plt.plot(ages, dev_salaries, label = "All_Devs", linestyle = '--', color = "black")
plt.title("Median Salaries in [USD] by Age")
plt.xlabel("Age")
plt.ylabel("Salaries in USD")
plt.legend()
plt.tight_layout()
```

```python
# We can replace all the plt present in the code with ax. Title, xlabel and ylabel will have set_ as prefix
ages = data["Age"]
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]

fig, ax = plt.subplots()

ax.plot(ages, py_salaries, label = "Python")
ax.plot(ages, js_salaries, label = "JavaScript")
ax.plot(ages, dev_salaries, label = "All_Devs", linestyle = '--', color = "black")
ax.set_title("Median Salaries in [USD] by Age")
ax.set_xlabel("Age")
ax.set_ylabel("Salaries in USD")
ax.legend()
plt.tight_layout()
```

```python
# What if we want python and js plot seperated out from All_devs plot but on the same figure??
# We will create 2 axis and 1 fig

ages = data["Age"]
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]

fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols=1)

ax2.plot(ages, py_salaries, label = "Python")
ax2.plot(ages, js_salaries, label = "JavaScript")
ax1.plot(ages, dev_salaries, label = "All_Devs", linestyle = '--', color = "black")


ax1.set_title("Median Salaries in [USD] by Age")
ax1.set_xlabel("Age")
ax1.set_ylabel("Salaries in USD")
ax1.legend()

ax2.set_title("Median Salaries in [USD] by Age")
ax2.set_xlabel("Age")
ax2.set_ylabel("Salaries in USD")
ax2.legend()

plt.tight_layout()
```

- In the above plot, since we are on 1 figure together, we can remove the title from the below graph and also the xlabel from the above graph by simply deleting the ax2.set_title and ax1.set_xlabel

```python
# sharing the title and xlabel
ages = data["Age"]
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]

fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols=1)

ax2.plot(ages, py_salaries, label = "Python")
ax2.plot(ages, js_salaries, label = "JavaScript")
ax1.plot(ages, dev_salaries, label = "All_Devs", linestyle = '--', color = "black")


ax1.set_title("Median Salaries in [USD] by Age")
ax1.set_ylabel("Salaries in USD")
ax1.legend()

ax2.set_xlabel("Age")
ax2.set_ylabel("Salaries in USD")
ax2.legend()

plt.tight_layout()
```

```python
# If we want x-coordinate values only on the second axis, we can do that by passing sharex argument to the subplot method
# Note: sharey method is also available

ages = data["Age"]
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]

fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols=1, sharex = True)

ax2.plot(ages, py_salaries, label = "Python")
ax2.plot(ages, js_salaries, label = "JavaScript")
ax1.plot(ages, dev_salaries, label = "All_Devs", linestyle = '--', color = "black")


ax1.set_title("Median Salaries in [USD] by Age")
ax1.set_ylabel("Salaries in USD")
ax1.legend()

ax2.set_xlabel("Age")
ax2.set_ylabel("Salaries in USD")
ax2.legend()

plt.tight_layout()
```

```python
# What if we want the above 2 plots on 2 different figures? We create another fig and axis argument

ages = data["Age"]
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax2.plot(ages, py_salaries, label = "Python")
ax2.plot(ages, js_salaries, label = "JavaScript")
ax1.plot(ages, dev_salaries, label = "All_Devs", linestyle = '--', color = "black")


ax1.set_title("Median Salaries in [USD] by Age")
ax1.set_xlabel("Age")
ax1.set_ylabel("Salaries in USD")
ax1.legend()

ax2.set_title("Median Salaries in [USD] by Age")
ax2.set_xlabel("Age")
ax2.set_ylabel("Salaries in USD")
ax2.legend()

fig1.savefig("fig1.png")
fig2.savefig("fig2.png")

plt.tight_layout()
```
