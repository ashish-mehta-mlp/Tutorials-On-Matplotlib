## Video 7: Scatter Plots

- Scatter plots are great when we want to show a relationship between 2 sets of values to see if they are co-related

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
plt.scatter(x,y)
plt.tight_layout()
```

- Scatter plots are great to different trends, outliers etc.
- Since the points in our scatter plot are randomly distributed all over the place, we can say that there is no trend between the two variables.
- Hence the two values are not co-related.

### Customizing the scatter plots
1. Changing the size of the dots

```python
# changing size of dots
x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
plt.scatter(x,y, s = 200)
plt.tight_layout()
```

```python
# Change the color of the dots
x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
plt.scatter(x,y, s = 150, color = "green")
plt.tight_layout()
```

```python
#Changing the marker of the graph
x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
plt.scatter(x,y, s = 150, color = "green", marker = "X")
plt.tight_layout()
```

```python
# Giving edges and alpha(see through effect) to the markers
x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
plt.scatter(x,y, s = 150, c = "green", edgecolor = "black", linewidth = 2, alpha = 0.75)
plt.tight_layout()
```

- The colors and sizes of the markers are on a per marker basis. 
- Using different colours and sizes of markers allows us to introduce different datasets on the same plot.

### Using differnt colors for markers.
- Lets pretend that our current data is some kind of a survey data and we want to further break down the data into something more specific
- Lets say that all these people have rated a product between 1-10 and we have to plot their ratings as well.
- For this we can assign different numbers to different possibilities and that will give you different colours. 

```python
x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
colors = [7,5,3,9,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5]
plt.scatter(x,y, s = 150, c = colors, cmap = "Reds", edgecolor = "black", linewidth = 2, alpha = 0.75)
plt.tight_layout()
# Cmap stands for color maps 
```

- Because of cmaps we get colors of differnt intensity
- Lighter colors are closer to zero
- Darker colors are closer to one

- But we have to give label to our colormap so that the viewers can understand what these color mean

```python
#Setting a colorbar legend
x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
colors = [7,5,3,9,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5]
plt.scatter(x,y, s = 150, c = colors, cmap = "Reds", edgecolor = "black", linewidth = 2, alpha = 0.75)
cbar = plt.colorbar()
cbar.set_label("Satisfaction Ratings")
plt.tight_layout()
```

- Along with color, we can also change the size of the datapoints aswell. This will help us to explain our data even further.
- Eg: Size of the scatter plot can be used for population explainations

```python
x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
colors = [7,5,3,9,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5]
sizes = [209,486,381,255,717,315,175,228,174,592,293,399,255,525,154,253,475,457,214,253]
plt.scatter(x,y, s = sizes, c = colors, cmap = "Reds", edgecolor = "black", linewidth = 2, alpha = 0.75)
cbar = plt.colorbar()
cbar.set_label("Satisfaction Ratings")
plt.tight_layout()
```

### Scatter plot for a CSV file data

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
data =  pd.read_csv("C:/Users/Ashish/Desktop/Python Tutorials/CSV files/yt.csv")
```

```python
data.head()
```

- The above data is a Youtube data. 
- First column is the number of views.
- Second column is the numver of likes.
- Third column is the ratio of likes vs dislikes

```python
# We have to see if the number of likes increase with number of views

views = data["view_count"]
likes = data["likes"]
ratio = data["ratio"]

plt.scatter(views, likes, edgecolor = "black", linewidth = 2, alpha = 0.75)

plt.title("Trending Youtube videos")
plt.xlabel("Number of views")
plt.ylabel("Number of likes")

plt.tight_layout()
```

- There are 200 points in total
- In the above plot we can see that there is one big outlier. Also, the points look very conjusted. Hence we make use of log scale 

```python
views = data["view_count"]
likes = data["likes"]
ratio = data["ratio"]

plt.scatter(views, likes, edgecolor = "black", linewidth = 2, alpha = 0.75)

plt.title("Trending Youtube videos")
plt.xlabel("Number of views")
plt.ylabel("Number of likes")

plt.xscale("log")
plt.yscale("log")
plt.tight_layout()
```

```python
# Finally, we can use the like/dislike ratio as the color parameter to better explain the data

views = data["view_count"]
likes = data["likes"]
ratio = data["ratio"]

plt.scatter(views, likes, c = ratio, cmap = "summer", edgecolor = "black", linewidth = 2, alpha = 0.75)

cbar = plt.colorbar()
cbar.set_label("Like to Dislike Ratio")

plt.title("Trending Youtube videos")
plt.xlabel("Number of views")
plt.ylabel("Number of likes")

plt.xscale("log")
plt.yscale("log")
plt.tight_layout()
```

## Video 8: Plotting Time Series Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mpl_dates
from datetime import datetime, timedelta
```

```python
plt.style.use("seaborn")
```

```python
date = [
    datetime (2019,5,24),
    datetime (2019,5,25),
    datetime (2019,5,26),
    datetime (2019,5,27),
    datetime (2019,5,28),
    datetime (2019,5,29),
    datetime (2019,5,30)
]

y = [0,1,3,4,6,5,7]

plt.plot_date(date, y)
```

- By Default the graph looks like a scatterplot. But we can easily make it a line plot by using **linestyle = "solid"** 

```python
date = [
    datetime (2019,5,24),
    datetime (2019,5,25),
    datetime (2019,5,26),
    datetime (2019,5,27),
    datetime (2019,5,28),
    datetime (2019,5,29),
    datetime (2019,5,30)
]

y = [0,1,3,4,6,5,7]

plt.plot_date(date, y, linestyle = "solid")
```

- Formatting the plot to make it look better:
    1. Auto formatting the coordinates of x-axis

```python
date = [
    datetime (2019,5,24),
    datetime (2019,5,25),
    datetime (2019,5,26),
    datetime (2019,5,27),
    datetime (2019,5,28),
    datetime (2019,5,29),
    datetime (2019,5,30)
]

y = [0,1,3,4,6,5,7]

plt.plot_date(date, y, linestyle = "solid")
plt.gcf().autofmt_xdate() # gcf (get current figure) autofmt(auto format)
plt.tight_layout()
```

2. Changing the format of dates 
   - From YYYY-MM-DD to DD-MM-YYYY
   - The module used in this case is mpl_dates

```python
date = [
    datetime (2019,5,24),
    datetime (2019,5,25),
    datetime (2019,5,26),
    datetime (2019,5,27),
    datetime (2019,5,28),
    datetime (2019,5,29),
    datetime (2019,5,30)
]

y = [0,1,3,4,6,5,7]

plt.plot_date(date, y, linestyle = "solid")
plt.gcf().autofmt_xdate() 
date_format = mpl_dates.DateFormatter("%d, %b,%Y")
plt.gca().xaxis.set_major_formatter(date_format) # gca = get current axis
plt.tight_layout()
```

- The date has been formatted to dd-mm-yyyy format

### Using the datetime Plot on CSV file

```python
data = pd.read_csv("C:/Users/Ashish/Desktop/Python Tutorials/CSV files/datetime.csv")
```

```python
data.head()
```

```python
price_date = data["Date"]
price_close = data["Close"]
```

```python
plt.plot_date(price_date, price_close, linestyle = "solid")
plt.gcf().autofmt_xdate()
plt.title("Bitcoin Prices")
plt.xlabel("Dates")
plt.ylabel("Price in (USD)")
```

- In the above graph, the x-axis is not plotted as dates but it has been plotted as strings

- To convert the string as datetime, we will be using a pandas method to set the string as datetime format on x-axis
- We can also sort the out of order dates to ordered dates

```python
# We are making use of pandas to_datetime method
data["Date"] = pd.to_datetime(data["Date"])
data.sort_values('Date', inplace = True)
price_date = data["Date"]
price_close = data["Close"]
plt.plot_date(price_date, price_close, linestyle = "solid")
plt.gcf().autofmt_xdate()
plt.title("Bitcoin Prices")
plt.xlabel("Dates")
plt.ylabel("Price in (USD)")
```
